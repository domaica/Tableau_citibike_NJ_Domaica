# TRANSFORM DATA

```python
# READ MERGED NEW CSV
df = pd.read_csv("Final_resources_NJ/2019/JC2019/ALL2019-citibike-NJ.csv")
df.head()
```

```python
# Elimination of mistakes or real trips not happened. Delete rows when trip duration is less than 90 seconds and 
# where start and end stations are the same
# Filter out records where users rented a bike but found that it wasn't working properly
df = df.loc[((df["tripduration"]>90) & (df["start station id"] != df["end station id"])),:]
df.head()
```

```python
# Convert trip duration column from seconds to minutes
df["tripduration"] = [trip/60 for trip in df["tripduration"]]

# Filter out trips over 24 hours long
df = df.loc[df["tripduration"] < 1440, :]

```

```python
# Rename trip duration column
df = df.rename(columns={"tripduration":"Trip Duration (min)"})

df.head()
```

```python
# Rename values in gender column from 0, 1, 2 which are meaningless to male and female
gender = {0:"Unknown", 1:"Male", 2:"Female"}
df.gender = [gender[item] for item in df.gender]
df.head()
```

```python
# Add ride id column to uniquely identify each ride
df["Ride ID"] = df["start station id"].map(str) + "_" + df["end station id"].map(str)
df.head()
```

```python
df = df.rename(columns={"birth year": "Birth_Year"})
```

```python
for col in df.columns:
    print(col)
```

```python
#change data types
df['starttime'] = pd.to_datetime(df['starttime'])
df['stoptime'] = pd.to_datetime(df['stoptime'])
```

```python
df.dtypes
```

```python
# Calculate age of riders from year where historically is collected
# empty list to save ages
ages = []
# loop to calculate age of riders and fill list
for i in df["Birth_Year"]:
    age = 2019 - i
    ages.append(age)
```

```python
# ages
```

```python
# df
```

```python
# add list to dataframe as column
df['age'] = ages
df
```

```python
# Delete all rows where user age is over 100
df = df.loc[df["age"] < 100]
df
```

```python
# Divide users in age bins
bins = [0, 17, 25, 35, 50, 65, 100]
labels = ["0-17", "18-25", "26-35", "36-50", "51-65", "65+"]

df["age group"] = pd.cut(df["age"], bins, labels=labels)
```

## LOAD

```python
# save dataframe to csv as loading data for tableau
df.to_csv("Final_resources_NJ/2019/FINALALL2019-citibike-NJ.csv", index=False)
```
